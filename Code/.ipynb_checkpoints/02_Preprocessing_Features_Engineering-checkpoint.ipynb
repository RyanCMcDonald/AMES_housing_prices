### Step 1: Reading the Data 
**Imports/ Read The Data**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../datasets/project_data/clean_train.csv")
pd.set_option('display.max_columns', 80)
train.head(1)
```

```python
train.shape
```

### Step 2: Feature Selection


#### Dummy Columns for better analysis in the models
  -**Transforming all 'Object-Type' data into dummy columns for better interpretation into models**

```python
train_object = train.select_dtypes(include=['object'])
train_object.head()
```

```python
train_object.shape
```

```python
features = train.loc[ : , train_object.columns] 
dummies = pd.get_dummies(data=features, drop_first=True)
dummies.head(3)
```

```python
train=pd.concat([train, dummies], axis=1)
train.shape
```

```python
train.isnull().sum().sort_values(ascending = False).head()
```

```python

train.dtypes.head()
```

```python
train_objects = train.select_dtypes(['object'])
train_objects
```

#### Dropping 'Object' dtype columns to allow correlation heatmap/analysis. Dummy columns to remain

```python
train.drop(columns=['ms_zoning','street','lot_shape','land_contour','lot_config','land_slope','neighborhood','condition_1','condition_2','bldg_type','house_style','roof_style','roof_matl','exterior_1st','exterior_2nd','mas_vnr_type','exter_qual','exter_cond','foundation','bsmt_qual','bsmt_cond','bsmt_exposure','bsmtfin_type_1','bsmtfin_type_2','heating','heating_qc','central_air','electrical','kitchen_qual','functional','garage_type','garage_finish','garage_qual','garage_cond','paved_drive','sale_type'], inplace=True)
```

```python
train.shape
```

#### Feature Engineering - New Columns

**One columns with all bathrooms combines**

**New columns for total finished square feet**

**Age of home column for better interpretation**

```python
train['total_baths']= ((train['bsmt_full_bath']+ train['full_bath']) + (train['bsmt_half_bath'] + train['half_bath'])* 0.5)

train['tot_fin_sq'] =train['bsmtfin_sf_1']+ train['bsmtfin_sf_2']+ train['gr_liv_area']

train['home_age'] = train['yr_sold']- train['year_built']
```

**Correlation Heatmap**

```python
plt.figure(figsize=(8,140))
ax = plt.axes()
ax.set_title('Numeric Feature Correlation to Sales Price')
corr = train.corr()[['saleprice']]
sns.set(font_scale=1)
sns.heatmap(corr.sort_values(by = 'saleprice', ascending = True), annot=True, cmap='mako', vmin = -1, vmax = 1, linecolor= 'black', linewidth = 2);
```
